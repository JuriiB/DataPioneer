```python
# Преобразование категориальных признаков
cat_col_names = ['порода', 'тип пастбища', 'порода папы_быка', 'возраст']
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder.transform(X_test[cat_col_names])

# Преобразование количественных признаков
num_col_names = ['эке', 'сырой протеин, г', 'спо']
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_test_scaled = scaler.transform(X_test[num_col_names])
```

```python
# Объединение преобразованных признаков
X_train = np.concatenate((X_train_ohe, X_train_scaled), axis=1)
X_test = np.concatenate((X_test_ohe, X_test_scaled), axis=1)

# Обучение модели 
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
predictions1 = model_lr.predict(X_test)
```

```python
# Проверка качества модели
mse = mean_squared_error(y_test, predictions1)
mae = mean_absolute_error(y_test, predictions1)
rmse = np.sqrt(mse)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)

r2 = r2_score(y_test, predictions1)  
print ("R2:", r2)
```

**Метрики моделей классификации, в том числе логистической регрессии, важны при анализе рисков или экономических расчётов**
- Точность (accuracy) - это мера того, насколько точно модель предсказывает класс объекта. Она рассчитывается как отношение числа правильно классифицированных объектов к общему числу объектов.
- Чувствительность (sensitivity) - это мера способности модели правильно классифицировать объекты одного класса. Она рассчитывается как отношение числа правильно классифицированных объектов одного класса к общему числу объектов этого класса.
- Lift chart - это графическое представление отношения числа правильно классифицированных объектов к общему числу объектов при различных порогах классификации.
- Precision - это мера точности предсказаний модели. Она рассчитывается как отношение числа правильно классифицированных объектов одного класса к общему числу предсказанных объектов этого класса.
- Recall - это мера полноты предсказаний модели. Она рассчитывается как отношение числа правильно классифицированных объектов одного класса к общему числу объектов этого класса.
FPR (False Positive Rate) - это мера вероятности ложноположительных результатов. Она рассчитывается как отношение числа неправильно классифицированных объектов другого класса к общему числу объектов другого класса.
Все эти метрики могут быть полезны при анализе рисков или экономических расчётов, так как они позволяют оценить качество модели классификации и ее способность правильно классифицировать объекты разных классов.
